```python
def balans():
    """ Beräknar saldot på kontot
    

    :return: saldot
    """  
    balans = 0
    for t in transaktioner:
        balans += t
    return balans

def validate_int(output, error_mess):
    while True:
        try:
            value = int(input(output))
            break
        except:
            print(error_mess)
    return value

def print_transactions():

    line = 0
    balans = 0
    output = ("\nAlla transaktioner:"
            "\n{:>3} {:>12} {:>12}"
            "\n______________________________").format("Nr", "Händelse", "Saldo")
    for t in transaktioner:
        line += 1
        balans += t
        output += ("\n{:>2}. {:>9} {:>9} kr".format(line, t, balans))

    return output

saldo = 1000
transaktioner = []
transaktioner.append(1000)

while True:
    menu = ("\n" 
            "\n Swedbank deluxe"
            "\n Saldo: {} kr" 
            "\n 1. Saldo" 
            "\n 2. Insättning" 
            "\n 3. Uttag" 
            "\n 4. Avsluta".format(balans()))

    val = validate_int(menu, "Felaktig inmatning!")

    if val == 1:
        print(print_transactions())

    elif val == 2:
        insättning = validate_int("Hur mycket vill du lägga in på kontot?", "Felaktig inmatning!")
        if insättning > 0:
            transaktioner.append(insättning)
        else:
            print("Måste vara större än 0")
    elif val == 3:
        uttag = validate_int("Hur mycket vill du dra ut?", "Felaktig inmatning!")
        if uttag <= balans() and uttag >= 0:
            transaktioner.append(-uttag)
        elif uttag < 0:
            print("Måste vara större än 0")
        else:
            print("Du har inte tillräckligt med pengar på kontot, fattiglapp")
    elif val == 4:
        break
    else:
        print("Fel input")

print("Adjöken fröken/herrn")
```
